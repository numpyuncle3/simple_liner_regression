```python
#使用tensorflow.keras进行简单的线性回归
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#这里依然使用random随机生成100组数据
data = []
for i in range(100):
    x = np.random.uniform(0.,100.)
    eps = np.random.normal(0,5)
    y = x*3.7 + 6 + eps
    data.append([x,y])
data_x = []
data_y = []
for i in range(len(data)):
    data_x.append(data[i][0])
    data_y.append(data[i][1])
print(data_x)
```

```python
plt.scatter(data_x,data_y)
```

```python
model = tf.keras.Sequential()
```

```python
print(data)
```

```python
#Dense层就是   ax+b
model.add(tf.keras.layers.Dense(1,input_shape= (1,)))
```

```python
model.summary()
```

```python
#模型的编译，可以理解为模型的配置 optimizer-->学习方法，loss-->损失
model.compile(optimizer = 'adam',loss = 'mse')
```

```python
#模型的训练
history = model.fit(data_x,data_y,epochs=5000)
```

```python
data_x_pre=[1,2,3]
model.predict(data_x_pre)
```
